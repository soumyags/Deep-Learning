1. Reading a coloured image


By Default any image is read as BGR(Blue,Green,Red in cv2)

```python
import cv2
img1=cv2.imread('/content/CuteCat.jpeg')

img1
```

```python
img2=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2
```

Seperate RGB Arrays

```python
R,G,B=cv2.split(img2)
```
